````python
import random

"""
social_story_coding_puzzle
- Make a Puzzle
- Evaluate Solution to Puzzle



"""


story_data_state_dictionary = {
    'a': {'absorbed': False,
                 'inventory': [],
                 'element': 'forest',
                 'energy': 1,
                 },
    'b': {'absorbed': False,
                 'inventory': [],
                 'element': 'forest',
                 'energy': 1,
                 },
    'c': {'absorbed': False,
                 'inventory': [],
                 'element': 'forest',
                 'energy': 1,
                 },
    'd': {'absorbed': False,
                 'inventory': [],
                 'element': 'forest',
                 'energy': 1,
                 },
    'e': {'absorbed': False,
                 'inventory': [],
                 'element': 'forest',
                 'energy': 1,
                 },
    'monsters_remaining': 1,
    'treasure_remaining': 1,
    'players_remaining': 5,
    }



def write_room_npc_item(room_array, item, row, col):
    """
    Write an item to the specified row and column in the room_array.
    """
    room_array[row][col] = item


def find_room_npc_item_location(room_array, item_name):
    """
    Find the specified item in
    the room_array and return its row and column.
    """
    for i, row in enumerate(room_array):
        for j, cell in enumerate(row):
            if item_name in cell:
                return i, j
    return None


def find_room_npc_item_location(room_array, letter):
    """
    Count the number of items in the
    room_array that start with the specified letter.
    """
    count = 0
    for row in room_array:
        for cell in row:
            if cell.startswith(letter):
                count += 1
    return count


def get_room_npc_items_list(room_array, letter):
    """
    Return a list of items in the room_array
    that start with the specified letter.
    """
    items = []
    for row in room_array:
        for cell in row:
            if cell.startswith(letter):
                items.append(cell)
    return items


def generate_npc_items(room_array, number_of_items, letter):
    """
    Generate the specified number of items starting with the specified letter and write them to random empty places in the room_array.
    """
    elements = ['fire', 'water', 'ice', 'forest', 'void']
    empty_cells = [(i, j) for i in range(len(room_array)) for j in range(len(room_array[0])) if room_array[i][j] == '']
    items = set()
    while len(items) < number_of_items:
        cell = random.choice(empty_cells)
        empty_cells.remove(cell)
        element = random.choice(elements)
        item = f'{letter}{len(items)+1},{element},0'
        if item not in items:
            items.add(item)
            room_array[cell[0]][cell[1]] = item


def update_player_status(player_pieces_data_list, story_data_state_dictionary, room_array, meta_data=None):
    """Updates the game state dictionary based on player piece information.

    Args:
        player_pieces_data_list (list): A list of strings, where each string represents
                             a player piece and its attributes in the format:
                             "<player_id>,<element>,<energy>".
        story_data_state_dictionary (dict): A dictionary representing the current state of the game.
                           This dictionary contains information about each player
                           (a, b, c, d, e), monsters, treasure, and remaining players.

    Returns:
        dict: The updated game state dictionary reflecting the player piece information.

    Example use:
        updated_story_data_state_dictionary = update_player_status(
            player_pieces_data_list,
            story_data_state_dictionary)
    """
    """
    # Example use:
    story_data_state_dictionary = {'a': {'absorbed': False, 'inventory': [], 'element': None, 'energy': 1},
                'b': {'absorbed': False, 'inventory': [], 'element': None, 'energy': 1},
                'c': {'absorbed': False, 'inventory': [], 'element': None, 'energy': 1},
                'd': {'absorbed': False, 'inventory': [], 'element': None, 'energy': 1},
                'e': {'absorbed': False, 'inventory': [], 'element': None, 'energy': 1},
                'monsters_remaining': 1, 'treasure_remaining': 1, 'players_remaining': 5}

    player_pieces_data_list = ['e,forest,1', 'a,ice,1', 'b,water,1', 'c,fire,1', 'd,void,1']

    updated_story_data_state_dictionary = update_player_status(player_pieces_data_list, story_data_state_dictionary)

    # Print the updated game state
    updated_story_data_state_dictionary
    """
    """
    TODO:
    maybe allow name-specification (Not just a b c)
    """

    for piece in player_pieces_data_list:
        player_id, element, energy = piece.split(',')  # Parse player piece string

        # Update player information in the game state
        story_data_state_dictionary[player_id]['element'] = element  # Set player's element
        story_data_state_dictionary[player_id]['energy'] = int(energy)  # Set player's energy


    # replace coma separated data with only player name string (a,b,c,d or e)
    # Iterate through each row of the array
    for row_index in range(len(room_array)):
        # Iterate through each cell in this row
        for col_index in range(len(room_array[row_index])):
            cell_data = room_array[row_index][col_index]
            # Check if the cell contains data
            if cell_data:
                # Split the data by comma
                player_id, _, _ = cell_data.split(',')
                # Update the cell with only the player ID
                room_array[row_index][col_index] = player_id

    return story_data_state_dictionary, room_array


def make_dungeon_room(instruction_dict=None):
    """
    Creates a random dungeon room array
    with player pieces placed strategically,
    and updates story_data_state_dictionary with character-elemnents

    Args:
        instruction_dict (dict, optional): Currently unused. Defaults to None.

    Returns:
        tuple: A tuple containing:
            - room_array (list): A 10x10 array representing the dungeon room.
            - room_metadata_dict (dict): Metadata about the room, including door location.

    Rules for room generation:
        - 10x10 array
        - Bottom row holds player pieces (a-e), each with a unique element and 1 energy unit.
        - Players are spaced randomly with at least one empty space between them.
        - All 5 elements (water, forest, ice, void, fire) must be present.

    Metadata includes:
        - door_location: Randomly chosen wall (top, right, bottom, left).

    input: maybe an instruction dict, maybe None

    output: a dungeon room array, a room_metadata_dict

    a python function to make a random dungeon room array
    according to a few rules:
    1. if the instruction dictionary is None, follow these rule.
    2. 10x10 array
    3. The bottom row holds, in random order, the player pieces
       one-space appart, e.g.
       test_room_array = [
        ...
        ['a,water,1', '', 'b,forest,1', '', 'c,ice,1', '', 'd,void,1 ', '', 'e,fire,1', '']
        ]
        or
        ['', 'a,void,1', '', 'b,forest,1', '', 'e,ice,1', '', 'd,water,1 ', '', 'c,fire,1']

        Each player must have a letter a,b,c,d,e with no duplicates
        (the letter is the character's unique id)
        and each player must have a ,1 meaning 1-energy-unit.

        But the sequence of the character and who has which element should be
        random, granted that the 5 elements are, like a,b,c,d,e, all represented
        without deletion or duplication: but who has which should be random.
        And as long as there is at least one space between players, the locations
        on the bottom row should be random.


    In the dungeon_room metadata dict, include:
    1. Where the door is around the wall. E.g.
    imagine there is another meta-array that is x+2,y+2 in size,
    showing in theory halls running outside any wall,
    and where in that larger space the door may be.

    # Example usage
    room_array, room_metadata_dict = make_dungeon_room()
    print("Metadata:", room_metadata_dict)
    for row in room_array:
        print(row)

    """

    if instruction_dict:
        # for now, story data is the only instruction
        story_data_state_dictionary = instruction_dict

    # Create the 10x10 room array
    room_array = [[''] * 10 for _ in range(10)]

    # Create the player pieces
    elements = ['water', 'forest', 'ice', 'void', 'fire']
    random.shuffle(elements)
    player_pieces_data_list = [f"{chr(97+i)},{elements[i]},1" for i in range(5)]
    random.shuffle(player_pieces_data_list)

    print(f"player_pieces_data_list -> {player_pieces_data_list} {type(player_pieces_data_list)}")

    # Place the player pieces in the bottom row with at least one space between
    valid_placement = False
    while not valid_placement:
        available_spaces = list(range(10))
        random.shuffle(available_spaces)
        placement_attempt = [""] * 10
        try:
            for piece in player_pieces_data_list:
                start_index = available_spaces.pop(0)
                placement_attempt[start_index] = piece
                # Remove adjacent spaces to ensure at least one empty space
                if start_index + 1 in available_spaces:
                    available_spaces.remove(start_index + 1)
                if start_index - 1 in available_spaces:
                    available_spaces.remove(start_index - 1)
            valid_placement = True
        except IndexError:  # Retry if placement fails due to not enough spaces
            print("Placement failed, trying again...")
            valid_placement = False

    room_array[-1] = placement_attempt  # Place the valid arrangement in the room


    """
    The door is designated by where the door is inside the room.
    """
    # Create the metadata dictionary
    door_location = random.choice(["top", "right", "bottom", "left"])
    door_relative_position = {
        "top": (1, random.randint(1, 10)),
        "right": (random.randint(1, 10), 10),
        "bottom": (10, random.randint(1, 10)),
        "left": (random.randint(1, 10), 1)
    }
    room_metadata_dict = {
        "door_location": door_location,
        "door_relative_position": door_relative_position[door_location]
    }


    # # TODO maybe do this in a separate function//area
    # # Update the story_data_state_dictionary with elements from room_array
    # story_data_state_dictionary, room_array = update_player_status(
    #     player_pieces_data_list,
    #     story_data_state_dictionary,
    #     room_array)


    return room_array, room_metadata_dict, story_data_state_dictionary


def calculate_cell_resonance_map(current_room_array):
    """
    Outputs a resonance report, detailing the resonance effect on each object
    based on neighboring elemental objects.

    Args:
        current_room_array (list): A 2D array representing the dungeon room.

    Returns:
        dict: A dictionary where keys are object locations (tuples) and values are
              the calculated resonance effect (positive, negative, or neutral).

    outputs a resonance report,
    of what the resonance effect is of each object
    based on whether there is another elemental object
    in the neighoring 8 squares,

    design note: this could be either in terms of
    the unique object name,
    or perhaps better the location of the object
    as names may not be unique

    possibly put in a taylored dungeon room array with just
    elements, as the full dungeon array may be both cluttered
    and evolving
    """
    print("\nstarting calculate_cell_resonance_map()")

    print('current_room_array -> ')
    for i in current_room_array:
        print(i)



    resonance_report = {}
    size = len(current_room_array)

    element_interactions = {
        'water': {'helps': 'forest', 'harms': 'fire'},
        'forest': {'helps': 'fire', 'harms': 'void'},
        'fire': {'helps': 'void', 'harms': 'ice'},
        'void': {'helps': 'ice', 'harms': 'water'},
        'ice': {'helps': 'water', 'harms': 'forest'},
    }

    for row in range(size):
        for col in range(size):
            cell = current_room_array[row][col]

            # # inspection
            # print(f"Cell ({row}, {col}): {cell}")  # Debugging: Print the cell being examined

            if cell:  # If cell is not empty
                # Extract only the element from the cell string
                object_element = cell.split(',')[1] if ',' in cell else cell

                # # inspection
                print(f"Object Element: {object_element}")  # Debugging: Print the extracted element

                has_positive_resonance = False  # Flag for positive resonance
                has_negative_resonance = False  # Flag for negative resonance

                # Check neighboring cells for resonance interactions
                for r in range(max(0, row - 1), min(size, row + 2)):
                    for c in range(max(0, col - 1), min(size, col + 2)):
                        if (r, c) != (row, col):  # Exclude the current cell
                            neighbor_cell = current_room_array[r][c]
                            # Extract only the element from the neighbor cell string
                            neighbor_element = neighbor_cell.split(',')[1] if ',' in neighbor_cell else neighbor_cell

                            # # inspection
                            print(f"Neighbor ({r}, {c}): {neighbor_element}")  # Debugging: Print the neighbor

                            if object_element and neighbor_element:
                                if neighbor_element == element_interactions[object_element]['helps']:
                                    has_positive_resonance = True
                                elif neighbor_element == element_interactions[object_element]['harms']:
                                    has_negative_resonance = True

                # Determine overall resonance effect
                if has_positive_resonance:
                    resonance_report[(row, col)] = 'positive'
                elif has_negative_resonance:
                    resonance_report[(row, col)] = 'negative'
                else:
                    resonance_report[(row, col)] = 'neutral'

                # print(f"Resonance Report for ({row}, {col}): {resonance_report[(row, col)]}")  # Debugging: Print the calculated resonance

    return resonance_report


# def setup_story_data_state_dictionary(story_data_state_dictionary, player_pieces_data_list, room_array, room_metadata_dict):
#     """
#     read puzzle setup data to create
#     1. set player 'names'
#     2. set player elements

#     ?. ?Set player energy starting levels?
#     """


#     # TODO maybe do this in a separate function//area
#     # Update the story_data_state_dictionary with elements from room_array
#     story_data_state_dictionary = update_player_status(
#         player_pieces_data_list,
#         story_data_state_dictionary)

#     return story_data_state_dictionary


# def move_object(room_array, instructions):
#     """Moves objects within the room array based on instructions.

#     Args:
#         room_array: The 2D array representing the room.
#         instructions: A list of tuples, each with the format (object_id, from_index, to_index).
#     """

#     for object_id, from_index, to_index in instructions:
#         # Find the row containing the object
#         for row_index, row in enumerate(room_array):
#             if object_id in row:
#                 from_col_index = row.index(object_id)
#                 # Remove object from current position
#                 room_array[row_index][from_col_index] = ''
#                 # Place object at new position
#                 room_array[row_index][to_index] = object_id
#                 break  # Stop searching once the object is moved


# def get_player_turn(turn_number_counting_from_1, number_of_players=5):
#     """
#     Determines the player whose turn it is based on the given turn number,
#     using % or 'mod' which gives the remainder of division, which
#     in this case is the same as the player's sequence.
#     with the exception that zero remainder means the last player.

#     Args:
#         turn_number_counting_from_1 (int): The current turn number.

#     Returns:
#         int: The player number (1 to 5) whose turn it is.
#     """
#     # Calculate the player number based on the turn number
#     player_number = turn_number_counting_from_1 % number_of_players

#     # If the remainder is 0, it means the last player's turn
#     if player_number == 0:
#         player_number = number_of_players

#     return player_number


def move_object(room_array, this_solution_action_list):
    """
    csv vs. id
    first item in csv is id
    if from, to and csv_id all match
    move csv to the move_to_coordinates
    remove the csv object from the move_from_coordinates location,
    replace with ''

    the from and to coordinates will be in tuples likes this:

    (object, row, column)

    move_from_coordinates = ('a',9,0)
    move_to_coordinates = ('a',9,1)

    a is the object in the array
    9 is the row
    0 is the column

    object, row, column


    # examples
    room_array = [
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['', '', '', '', '', '', '', '', '', ''],
    ['a', '', 'c', '', 'd', '', 'e', '', 'b', '']
    ]

    """
    print(f"def move_object()")


    # get move-from-instruction id
    obj1, from_row, from_col = this_solution_action_list[0]

    # get move-to-instruction id
    obj2, to_row, to_col = this_solution_action_list[1]

    # get id in csv in room-array
    csv_id = room_array[from_row][from_col]

    # check-stuff and do-stuff
    if (obj1 == obj2) and (obj2 == csv_id):
        """
        if all id's match
        """

        # get data to move from the room-array
        csv_object = this_solution_action_list[0]

        # Place the object in the new location
        room_array[to_row][to_col] = csv_object

        # Remove the object from the original location
        room_array[from_row][from_col] = ''

    else:
        # Do nothing
        return room_array




def get_player_turn(turn_number_counting_from_1, number_of_players=5):
    """
    Determines the player whose turn it is based on the given turn number,
    using % or 'mod' which gives the remainder of division, which
    in this case is the same as the player's sequence.
    with the exception that zero remainder means the last player.

    Args:
        turn_number_counting_from_1 (int): The current turn number.

    Returns:
        int: The player number (1 to 5) whose turn it is.
    """
    # Calculate the player number based on the turn number
    player_number = turn_number_counting_from_1 % number_of_players

    # If the remainder is 0, it means the last player's turn
    if player_number == 0:
        player_number = number_of_players

    return player_number

def validate_move(
    this_move_number_from_0,
    this_move_tuple,
    story_data_state_dictionary,
    solution_meta_data,
    room_array,
    room_metadata_dict,
    turn_counter_from_1,
):
    """Validates a single move within a solution.

    Args:
        this_move_number_from_0 (int): The index of the move within the solution.
        this_move_tuple (tuple): A tuple representing the move, e.g., ('a', 9, 0).
        story_data_state_dictionary (dict): The current game state.
        solution_meta_data (dict): Metadata about the solution, including player turns and action types.
        room_array (list): The 2D array representing the room.
        room_metadata_dict (dict): Metadata about the room, including door location.

    Returns:
        bool: True if the move is valid, False otherwise.



    Todo:
    ? Maybe make sure that these two:
        solution_move_tuple_list = [
        solution_meta_data = {
    are in agreement about what piece is acting/moving

    """

    print("\nvalidate_move()")

    print(f"turn_counter_from_1 -> {turn_counter_from_1}")
    print(f"this_move_tuple -> {this_move_tuple}")

    # From
    from_this_move_tuple = this_move_tuple[0]

    # To
    to_this_move_tuple = this_move_tuple[1]

    player_id, from_row, from_col = from_this_move_tuple  # Unpack move tuple
    player_id, to_row, to_col = to_this_move_tuple  # Unpack move tuple

    # 1. Check if it is the player's turn
    move_name_lookup = {
        1: 'a',
        2: 'b',
        3: 'c',
        4: 'd',
        5: 'e',
    }


    adjusted_move_number = this_move_number_from_0 + 1  # Adjust for 1-based indexing in solution_meta_data

    expectedplayer_number = get_player_turn(adjusted_move_number)
    expected_player = move_name_lookup[expectedplayer_number]


    player_for_move = solution_meta_data[this_move_number_from_0]['player']  # Access correct dictionary key
    if player_for_move != expected_player:
        print(f"Move {adjusted_move_number} invalid: Not {player_for_move}'s turn. It's {expected_player}'s turn.")
        return False

    # Check if the move number is out of range
    if this_move_number_from_0 >= len(solution_meta_data):
        print(f"Move {this_move_number_from_0} invalid: Move number out of range.")
        return False

    # 3. check if move agrees between solution_move_tuple_list & solution_meta_data
    this_player_from_movetuple1 = this_move_tuple[0][0]
    this_player_from_movetuple2 = this_move_tuple[1][0]

    this_player_from_solution_metadata = solution_meta_data[this_move_number_from_0]["player"]
    print(f"this_player_from_movetuple1 -> {this_player_from_movetuple1}")
    print(f"this_player_from_movetuple2 -> {this_player_from_movetuple2}")
    print(f"this_player_from_solution_metadata -> {this_player_from_solution_metadata}")

    # check_1 = this_player_from_movetuple2 != this_player_from_solution_metadata
    check_2 = this_player_from_movetuple1 != this_player_from_movetuple2
    # check_3 = this_player_from_movetuple1 != this_player_from_solution_metadata

    # if not check_1 or not check_2 or not check_3:
    if not check_2:
        print(f"The player in question is not in agreement.")
        return False

    # 3. Check action type and delegate to specific validation function
    action_type = solution_meta_data[this_move_number_from_0]['action_type']
    if action_type == 'move_self':
        return validatemove_self(player_id, from_row, from_col, to_row, to_col, story_data_state_dictionary, room_array)
    elif action_type == 'throw':
        return validate_throw(player_id, from_row, from_col, to_row, to_col, room_array)
    elif action_type == 'exit_room':
        return validate_exit_room(player_id, from_row, from_col, story_data_state_dictionary, room_array, room_metadata_dict)
    else:
        print(f"Move {this_move_number_from_0} invalid: Unknown action type '{action_type}'.")
        return False





# log these TODO
def validatemove_self(player_id, from_row, from_col, to_row, to_col, story_data_state_dictionary, room_array):
    """Validates a 'move_self' action.
    modify for multi whatever in array ...
    or modify input?


    """
    from_room_array_list = room_array[from_row][from_col].split(',')

    from_room_array_character = from_room_array_list[0]
    from_room_array_element = from_room_array_list[1]
    from_room_array_energy = from_room_array_list[2]

    # print(f"""
    # from_room_array_character -> {from_room_array_character} {type(from_room_array_character)}
    # from_room_array_element -> {from_room_array_element} {type(from_room_array_element)}
    # from_room_array_energy -> {from_room_array_energy} {type(from_room_array_energy)}
    # """)


    # print(f"""
    # player_id-> {player_id} {type(player_id)}
    # from_row-> {from_row} {type(from_row)}
    # from_col -> {from_col} {type(from_col)}
    # to_row -> {to_row} {type(to_row)}
    # to_col -> {to_col} {type(to_col)}
    # story_data_state_dictionary -> {story_data_state_dictionary} {type(story_data_state_dictionary)}
    # room_array -> {room_array} {type(room_array)}

    # """)


    # 1. Check if player has enough energy
    if story_data_state_dictionary[player_id]['energy'] < 1:
        print(f"Move invalid: Player {player_id} has insufficient energy to move.")
        return False

    # 2. Check if the move is within the room bounds
    if not (0 <= to_row < len(room_array) and 0 <= to_col < len(room_array[0])):
        print(f"Move invalid: Move_to location ({to_row}, {to_col}) is out of bounds.")
        return False

    # 3. Check if the starting position is correct
    if from_room_array_character != player_id:
        print(f"Move invalid: Player {player_id} is not at starting location ({from_row}, {from_col}).")
        return False

    # 4. Check if the target location is empty
    if room_array[to_row][to_col] != '':
        print(f"Move invalid: Move_to location ({to_row}, {to_col}) is not empty.")
        return False

    # 5. Check if the move is to an adjacent square
    if abs(from_row - to_row) > 1 or abs(from_col - to_col) > 1:
        print(f"Move invalid: Player {player_id} can only move to an adjacent square.")
        return False

    return True


# TODO what is this empty?
def validate_throw(player_id, from_row, from_col, to_row, to_col, room_array):
    """Validates a 'throw' action."""
    # ... (Implementation for throw validation - similar structure to move_self)
    return True  # Replace with actual validation logic


def validate_exit_room(player_id, from_row, from_col, story_data_state_dictionary, room_array, room_metadata_dict):
    """Validates an 'exit_room' action."""
    # 1. Check if player has enough energy
    if story_data_state_dictionary[player_id]['energy'] < 1:
        print(f"Move invalid: Player {player_id} has insufficient energy to exit.")
        return False

    # 2. Get door location
    door_row, door_col = room_metadata_dict['door_relative_position']

    # 3. Check if the player is at the door location
    if from_row != door_row or from_col != door_col:
        print(f"Move invalid: Player {player_id} is not at the door location ({door_row}, {door_col}).")
        return False

    return True


def update_dungeon_room_array_one_turn(
        this_turn_action_tuple_list,
        story_data_state_dictionary,
        room_array,
        room_metadata_dict,
        turn_counter_from_1,
        this_index_in_solution,
):
    """
    current_room_array ~ 10x10 array:
    cells either empty, to occupied by a noun_code:

    letter + number for characters:
    e.g.
    a,ice,1 is player-1 with 1 energy-point
    a,water,2

    whisps: move left or right so
    l0 or r0 or l & r are the whisps
    or l,fire,0 or r,forest,0
    or ra, li, etc.


    iterate through the array from 0,0


    Step 1.
    resonance effects
    for each occupied cell

    Use helper-function to map out the resonance effect on each cell.
    the output is a dictionary where the effects can be updated:
    1. players will have their stats changed
    2. whisps will disappear if <0, and multiply if >0
    3.


    ### Steps for crawling through moves:
    1. resonance map
    2. resonance effect assign
    - update story_data_state_dictionary
    3. get moves/actions
    4. carry out moves/actions
    -
    5. detect of all monsters gone
    6. detect if treasure taken
    7. detect if all players have exited via the door
    8. detect if game over success or fail

    Check story_data_state_dictionary:

    ### Rules:
    1. Nonesense instructions will be ignored.
    2. actions fully stated: move piece from, move piece to

    # TODO:
    - what notation to use for

    - issues:
    whisp, treasure, energy



    actions:
    - move:
        move_tuple_example = ('pc4','pc5')
    - throw:
        throw_move_tuple = ('pc4','pc5')
    - remove item from inventory:
        (item_to_remove, empty_square_to_put_it_on)


    this_turn_action_tuple_list:

    based on chess




    if player obsorbed,

    action: remove item from inventory: player


    """
    print("\nupdate_dungeon_room_array_one_turn()")

    print("this move number?")

    # this_turn_action_tuple_list,
    # story_data_state_dictionary,
    # room_array,
    # room_metadata_dict

    print("room_array")
    print(room_array)
    for i in room_array:
        print(i)

    ########
    # Setup
    ########

    #
    ### Steps for crawling through moves:
    # 1. resonance map

    resonance_report = calculate_cell_resonance_map(room_array)

    print("resonance_report")
    print(resonance_report)


    # 2. resonance effect assign
    # - update story_data_state_dictionary

    def update_based_on_resonance_effects(
            resonance_report,
            room_array,
            story_data_state_dictionary):
        pass


    ############################
    # 3. validate moves/actions
    ############################
    # player_id, move_from_here, move_to_here = this_turn_action_tuple_list.split(',')
    print(f"this_turn_action_tuple_list -> {this_turn_action_tuple_list} {type(this_turn_action_tuple_list)}")



    # [('a',9,0),('a',9,1), ]
    # player_id, move_from_here, move_to_here = this_turn_action_tuple_list
    # print(f"player_id -> {player_id} {type(player_id)}")
    # print(f"move_from_here -> {move_from_here} {type(move_from_here)}")
    # print(f"move_to_here -> {move_to_here} {type(move_to_here)}")


    """
    Validate moves
    remove invalid moves


    future TODO
    check that each player takes an action only on their turn
    ok: action: throw adjacent object
    ok: action: move yourself

    make a move_and_type array
    for what character is making what type of action (move_self, throw, etc.)
    check that is happening on their turn
    check that throw or move self is consistent with their location
    on the map


    """


    #############################
    # 4. carry out moves/actions
    #############################
    """
    Make sure move is 'valid'
    e.g. if that piece is not on the move_from_here square
    or
    if the move_to_here square is not empty
    then the action is invalid: pass
    log invalid action:
    - move number
    - move
    - reason
        """
    ###################
    # validity_of_move
    ###################

    # preset/reset value
    valid_check_boolean = False

    # check
    valid_check_boolean = validate_move(
        this_move_number_from_0=this_index_in_solution,
        this_move_tuple=this_turn_action_tuple_list,
        story_data_state_dictionary=story_data_state_dictionary,
        solution_meta_data=solution_meta_data,
        room_array=room_array,
        room_metadata_dict=room_metadata_dict,
        turn_counter_from_1=turn_counter_from_1,
    )

    # The Moving!
    if valid_check_boolean is True:
        move_object(room_array, this_turn_action_tuple_list)

    # # 5. detect of all monsters gone
    monster_list = get_room_npc_items_list(room_array, letter='w')
    print(f"monster_list {monster_list} {type(monster_list)}")

    # # 6. detect if treasure taken
    treasures_list = get_room_npc_items_list(room_array, letter='t')
    print(f"treasures_list {treasures_list} {type(treasures_list)}")

    # # 7. detect if all players have exited via the door
    # ...

    # # 8. detect if game over success or fail
    # ...

    # move iterate through the array from 0,0
    # room_array, room_metadata_dict = make_dungeon_room()

    # setup_story_data_state_dictionary(story_data_state_dictionary, room_array, room_metadata_dict)



    # return updated dictionary
    return story_data_state_dictionary


def iterate_through_whole_game(
        solution_move_tuple_list,
        solution_meta_data,
        room_array,
        story_data_state_dictionary,
        room_metadata_dict):

    print("\niterate_through_whole_game()")

    # preset / reset
    pass_fail = False

    print(room_array)
    for  i in room_array:
        print(i)

    turn_counter_from_1 = 1

    for this_index_in_solution, this_turn_action_tuple_list in enumerate(solution_move_tuple_list):

        story_data_state_dictionary = update_dungeon_room_array_one_turn(
            this_turn_action_tuple_list,
            story_data_state_dictionary,
            room_array,
            room_metadata_dict,
            turn_counter_from_1,
            this_index_in_solution,
            )

        turn_counter_from_1 += 1

    # todo: finish this function

    #####################
    # Make Game Timeline
    #####################

    game_timeline = {}


    return pass_fail, game_timeline


####################
####################
# Configure and Run
####################
####################
"""
1. Configure: Pick a puzzle type
2. Make a puzzle
3. Make a solution
4. Test the solution
5. Look at the results
"""

###############
# 1. Configure
###############
"""
~ difficulty
~ mode
~ starting details
"""
print("\n\n1. Configure")



"""
update:
- monsters_remaining
- treasure_remaining
- players_remaining
"""
story_data_state_dictionary = {
    'turn_number_from_1': 1,
    'turn_number_from_0': 0,
    # 'players_turn': get_player_turn(turn_number, number_of_player:5),  # get_player_turn(turn_number, number_of_player:5)
    'room_array': [],
    'player_move_list': [], # dict?
    # validations:
    'piece_player_movements': [],  #

    'resonance_interactions': [],  # calculate_cell_resonance_map(current_room_array) -> resonance report

    'treasure_status': False,
    'left_room_status': False,
    'natural_language_turn_description': "",

    'monsters_remaining': 1,
    'treasure_remaining': 1,
    'players_remaining': 5,

    'a': {'absorbed': False,
                 'inventory': [],
                 'element': None,
                 'energy': 1,
                 },
    'b': {'absorbed': False,
                 'inventory': [],
                 'element': None,
                 'energy': 1,
                 },
    'c': {'absorbed': False,
                 'inventory': [],
                 'element': None,
                 'energy': 1,
                 },
    'd': {'absorbed': False,
                 'inventory': [],
                 'element': None,
                 'energy': 1,
                 },
    'e': {'absorbed': False,
                 'inventory': [],
                 'element': None,
                 'energy': 1,
                 },

    }


# vs.

# story_data_state_dictionary = {
#     'a': {'absorbed': False,
#                  'inventory': [],
#                  'element': None,
#                  'energy': 1,
#                  },
#     'b': {'absorbed': False,
#                  'inventory': [],
#                  'element': None,
#                  'energy': 1,
#                  },
#     'c': {'absorbed': False,
#                  'inventory': [],
#                  'element': None,
#                  'energy': 1,
#                  },
#     'd': {'absorbed': False,
#                  'inventory': [],
#                  'element': None,
#                  'energy': 1,
#                  },
#     'e': {'absorbed': False,
#                  'inventory': [],
#                  'element': None,
#                  'energy': 1,
#                  },
#     'monsters_remaining': 1,
#     'treasure_remaining': 1,
#     'players_remaining': 5,
#     }

print(story_data_state_dictionary)


###################
# 2. Make a puzzle
###################
print("\n\n2. Make a puzzle")
room_array, room_metadata_dict, story_data_state_dictionary = make_dungeon_room(story_data_state_dictionary)


print(f"Metadata: {room_metadata_dict}")
print("room_array:")
for row in room_array:
    print(row)

print(f"new story_data_state_dictionary (with elements): {story_data_state_dictionary}")


####################################
# 3. Validate This Move in Solution
####################################
"""
player_id, from, to = solution.split(',')

Check only one move for whether that move is valid in a
context of:

e.g.

def validate_move (
    this_move_number,
    this_move_tuple,
    story_data_state_dictionary,
    solution_meta_data,
    room_array,
    )

There are several issues that could invalidate a move:

note: players move in order a,b,c,d,e,a,b,c,...etc.
Pay attention to whether you are counting from 0 or 1.

assume:
1. NPCs/monsters/objects act on this turn before player characters
2.

for exit_room:
1. must be player's turn (see solution meta_data & number of move)
2. player must have 1 energy (see story_data_state_dictionary)
3. player must be on door-square

for throw:
1. must be player's turn (see solution meta_data)
2. object to throw must be next to you (8 neighbor squares)
3. throw target must be one move_from_here square
4. throw move_to_here location must be empty
5. structure of the move is not valid:
not two tuples: each with three parts: string int int

for move_self:
1. must be player's turn (see solution meta_data)
2. player must have 1 energy (see story_data_state_dictionary)
3. move_self target must be one move_from_here square
4. move_self move_to_here location must be empty
5. structure of the move is not valid:
not two tuples: each with three parts: string int int

It may be wise to create a separate do-one-thing-well
function for each type of validation operation.

"""

# examples
room_array = [
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['', '', '', '', '', '', '', '', '', ''],
['a,fire,1', '', 'c,ice,1', '', 'd,forest.1', '', 'e,void,1', '', 'b,water,1', '']
]

# Add NPC items
"""
Add n whisps as 'w'
add n teasure chests as 't'
"""
generate_npc_items(room_array, 5, 'w')
generate_npc_items(room_array, 3, 't')


solution_move_tuple_list = [
    [('a',9,0),('a',9,1), ],
    [('b',9,8),('b',9,7), ],
    [('a',9,1),('a',9,0), ],
    [('b',0,9),('b',1,2), ],
    ]

solution_meta_data = {
    0: {'player': 'a', 'action_type':'move_self'},
    1: {'player': 'b', 'action_type':'move_self'},
    2: {'player': 'c', 'action_type':'throw'},
    3: {'player': 'd', 'action_type':'exit_room'},
    }


#######################
# 4. Test the solution
#######################
pass_fail, game_timeline = iterate_through_whole_game(
        solution_move_tuple_list,
        solution_meta_data,
        room_array,
        story_data_state_dictionary,
        room_metadata_dict)


#########################
# 5. Look at the results
#########################
print(pass_fail, game_timeline)
````

# Modifying the Traditional five-element system

We are using a f-elment system because it is a wonderful model of differing cross-effects. Everything effects something else positively and something else negatively (and like-wise is affected positively and negatively by other things).

We are using a 'modified' version because a lot of the 'help' or 'harm' relationships in the original system do not make enough intutitive or rational sense in a concrete way for our purposes here. We are not studying abstract relationships in eastern medicine, we want concrete relationships for spells in a story game.

Arguably the best parts of the orignal system are that water harms fire, and wood helps fire. But most of the other relationships are either nonsense or arbitrary stretches of rationalization that could be argued in any direction.

Specifically, these are the dubious relationships from the original system.

## Relationships that do not make enough sense:
- Earth harms water (earth soil and rocks are integral to the water cycle)
- Wood harms earth (Wood in no way harms soil or rocks...)
- fire harms metal (Metal only exists as such because of fire.)
- metal harms wood (Metal in a forest will be hapily ignored.)

 &
- fire helps earth (You line a firepit with rocks because fire has no effect on the rocks)
- earth helps metal (Adding rocks and dirt to metal won't help the metal and will probably damage it.)
- metal helps water (Most metals are not something you want to add to your drinking water.)
- water helps wood (Water helps forests, but it is BAD for wood.)
